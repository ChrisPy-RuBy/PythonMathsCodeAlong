# **Sequences**

A *sequence* in mathematics is a list of numbers generated by a 'rule' or an 'algorithm' of some sort. There are different ways to define the rule. The first way is by specifying the *closed form* of the sequence  

**Closed form**

```python
# A sequence defined in closed form
from sympy import var, init_printing
init_printing()
var('n')
a_n = 4*n
```

So the closed form can be viewed, more-or-less, as a straightforward function: you provide n, or a range of n-values, and it returns the corresponding element(s) of the sequence, a_n. Here a_n would typically be pronounced *'a subscript n'* (or *'a sub n'*), and this is the convention indicating the n-th element of each sequence. Variable n is usually an integer.

```python
print([a_n.subs({n: i}) for i in range(0,10)])
```

Note that the closed form gives us, for want of a better word, *'direct'* access to any element of the sequence we are interested in. Say we want to calculate a_10: we obtain this directly as a_10 = 4\*10 = 40 without needing to calculate any other elements of the sequence. This seems merely a statement of the obvious, but compare this behaviour with the *recurrence relation* definition of a sequence discussed below, where direct access of this sort is not possible.  

SymPy provides a *sequence* class to define and manipulate sequences. Here's an example of use:

```python
from sympy import sequence, oo
```

```python
# we can define a sequence by formula in closed form
a_n = sequence(4*n, (n, 0, oo))
a_n.coeff(1), a_n.coeff(2), a_n.coeff(oo)
```

```python
# What type is a_n?
print(type(a_n))
a_n
```

```python
# The class supports slice indices
# Don't forget, as with ranges, the final index is *not* included
a_n[1:4]
```

```python
# Can also return the interval upon which the sequence is defined
a_n.interval
```

**Recurrence relation form**

The second way to define a sequence is in the form of a *recurrence relation*. This simply means that we calculate values for new elements of the sequence based upon previous elements. Here's a simple example:

```python
var('n')
def a_n(n):
    if n == 1: 
        return 1
    return a_n(n-1) + 4

a_n(n).subs({n: 3})
```
